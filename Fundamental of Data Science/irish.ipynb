```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris_df['target'] = iris.target
iris_df.head()
```

```python
iris_df['Flower Name'] = iris_df.target.apply(lambda x: iris.target_names[x])
iris_df.head()
```

```python
iris_df.info()
```

Statistics analysis

```python
iris_df.describe()
```

```python
numerical_variable = iris_df.select_dtypes(include=['int','float'])
variance = numerical_variable.var(ddof=1)
print(variance) 
```

```python
#DATAFRAME FOR SETOSA
df0 = iris_df[iris.target==0]
df0.head()
```

```python
#DATAFRAME FOR VERSICOLOR
df1 = iris_df[iris.target==1]
df1.head()
```

```python
#DATAFRAME FOR VIRGINICA
df2 = iris_df[iris.target==2]
df2.head()
```

Visualization

```python
# Sepal visualization
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'blue')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'green')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'red')
```

```python
# Petal visualization 
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'blue')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color = 'red')
```



X-axis (Sepal_Length_cm):
- The x-axis represents the sepal length in centimeters. The range appears to go from approximately 4.3 cm to 8.0 cm.

Y-axis (Count):
- The y-axis represents the count, or how many observations (or samples) fall within each bin (range of sepal lengths). For example, the tallest bar around 5.0 cm shows that this range has the most samples (about 16).

```python
# Histogram for Sepal Length
plt.figure(figsize=(10,7))
x = iris_df['sepal length (cm)']

plt.hist(x, bins = 20, color='blue', edgecolor = 'black')
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(10,7))
x = iris_df['sepal width (cm)']

plt.hist(x, bins = 20, color='blue', edgecolor = 'black')
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(10,7))
x = iris_df['petal length (cm)']

plt.hist(x, bins = 20, color='blue', edgecolor = 'black')
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.xticks(np.arange(min(x), max(x) + 0.5, 0.5))
plt.ylabel("Count")
```

```python
plt.figure(figsize=(10,7))
x = iris_df['petal width (cm)']

plt.hist(x, bins = 20, color='blue', edgecolor = 'black')
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.xticks(np.arange(min(x), max(x) + 0.5, 0.5))
plt.ylabel("Count")
```

## Density (PDF = Probability Density Function) Visualization

```python
sns.FacetGrid(iris_df, hue="Flower Name", height=5) \
   .map(sns.kdeplot, "petal length (cm)") \
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(iris_df, hue="Flower Name", height=5) \
   .map(sns.kdeplot, "petal width (cm)") \
   .add_legend()
plt.show()
```
